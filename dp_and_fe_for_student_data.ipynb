#Mean Replcing ค่าเฉลี่ย

**Standarization ข้อมูลที่เป็นมาตรฐาน**

เพศ มีข้อมูลอะไรบ้าง
เราจะเลือกจัดการค่าเพศแก้ให้มีค่ามาเหมือนกัน

```python
data['เพศ']
```

```python
def standardize_gender(name):
    if name in ['ชาย', 'Male', 'ช', 'M']:
        return 'ชาย'
    elif name in ['หญิง', 'Female', 'ญ', 'F']:
        return 'หญิง'
    else:
        return name

data['เพศ'] = data['เพศ'].apply(standardize_gender)
```

ค่าของเพศจะเป็นมาตรฐาน

```python
data
```

คณะ มีข้อมูลอะไรบ้าง เราจะเลือกจัดการคณะแก้ให้มีค่ามาเหมือนกัน

```python
data['คณะ']
```

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']: #วิทยาศาสตร์ - วิทย์ , Science ,Sci
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:#ศิลปศาสตร์ - Arts,ศิลป์ ,Liberal Arts
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์','วิศวกรรม','วิศวะ','Engineering','วิศวฯ']: #วิศวกรรมศาสตร์ - วิศวกรรม,วิศวะ ,Engineering ,วิศวฯ
        return 'วิศวกรรมศาสตร์'
    else:
        return name
data['คณะ'] = data['คณะ'].apply(standardize_faculty)
```

ค่าของคณะจะเป็นมาตรฐาน

```python
data
```

**Abnormaly Value การจัดการค่าผิดปกติ**

แทนค่าNaNเพื่อจะให้มีค่าโดยfillna

```python
data['อายุ'].fillna(data['อายุ'].mode()[0], inplace=True)
data['GPA'].fillna(data['GPA'].mode()[0], inplace=True)
data.isnull().sum()
```

```python
data
```

ขอบเขตของอายุ , Q1 IQR Q3

```python
Q1_age = data['อายุ'].quantile(0.25)
Q3_age = data['อายุ'].quantile(0.75)
```

```python
Q1_age, Q3_age
```

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {data['อายุ'].mean():.2f}")
```

```python
IQR_age = Q3_age - Q1_age
```

```python
IQR_age
```

หาขอบเขตอายุ ด้วยขอบเขตล่าง ขอบเขตบน

```python
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age
```

```python
lower_bound_age, upper_bound_age
```

```python
data['อายุ'] >= lower_bound_age
```

```python
data['อายุ'] <= upper_bound_age
```

อันนี้จะเอา5ออกจากทั้งหมด เพราะอายุเกิน

```python
data_no_outliers_age = data[(data['อายุ'] >= lower_bound_age) & (data['อายุ'] <= upper_bound_age)]
data_no_outliers_age
```

เกรดมีได้แค่1-4

```python
lower_bound_gpa = 1
upper_bound_gpa = 4
```

```python
data['GPA'] >= lower_bound_gpa
```

```python
data['GPA'] <= upper_bound_gpa
```

เอาคนที่เกรดผิดปกติออกไป

```python
data_no_outliers_gpa = data[(data['GPA'] >= lower_bound_gpa) & (data['GPA'] <= upper_bound_gpa)]
data_no_outliers_gpa
```

**หาค่าน้ำหนักBMI โดยการผสมสูตร**

```python
data['BMI'] = data['น้ำหนัก_kg'] / (data['ส่วนสูง_cm']/100)**2
```

```python
data['สุขภาพ'] = pd.cut(data['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```
